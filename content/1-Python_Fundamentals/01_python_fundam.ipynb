# II. Syntax, Variables, Data Types

## A. Syntax

Think of **Python syntax** as the *grammar* of the programming language. It describes:

* where things must go
* what symbols you must use
* how Python knows what you mean

If you break the rules ‚Üí **syntax error** üò¨

Syntax matters because every program you write has to do three things:

1. **Read input** ‚Äî Python needs to understand what you typed.
2. **Process** ‚Äî it follows your instructions to figure out what to do.
3. **Write output** ‚Äî it produces the result you asked for.

**TL;DR:** It's not enough to know the right *words* ‚Äî they have to be in the right *order* and the right *structure*. That structure is syntax, which includes things like indentation, colons, parentheses, and even cAsE sEnSiTiViTy.

And because the ‚Äúright place‚Äù often depends on the *type* of thing you‚Äôre writing (a variable, a function, a loop, etc.), we‚Äôll point out important syntax rules as we go.

Next up: **variables and data types**. üéâ

## B. Variables and Data Types

Quick overview: 
- String
- Integer
- Floating-Point Number
- Boolean

Let's jump right in to some mini-programs!

````python
# Variable and type demo 1
a = 'Hello World'
print(a)
print(type(a))
````

Here, we have a **variable** called `a`.   
When we write that it is "equal to 'Hello World'", we are telling the program to *store* something using `a`.   
*Variables* are what we use to store something.  

- `print()` is a function: a reusable block of code that performs a specific task. Whatever is between the parentheses `()` of a function is called an **argument**. (Parentheses! Syntax! üëÄ)

- `print(a)` is an instruction to output the value *assigned* to `a`. 
- `print(type(a))` asks for the *type* of data that this *assigned value* has. 

What `<class 'str'>` tells us is that `'Hello World'` is a **string**: the Python type used to store and manipulate **text**.  
- Whenever you place values between single quotes, it becomes a string value. (Single quotes! Syntax! üëÄ)  

````python
# example: 
var = 'This is a string.'
print(var)
````

üö® If you want to use an apostrophe/single quote INSIDE your string, replace the outside single quotes with double.

````python
# example: 
var = "Couldn't this also be a string?"
print(var)
print(type(var))
````

# III. Lists, Tuples, Sets, and Dictionaries

In Python, sometimes a single variable isn‚Äôt enough. What if you want to store **multiple values together**?

That‚Äôs where **collections** come in. Python has four main types:

* **List**
* **Tuple**
* **Set** 
* **Dictionary**

**Collections let us store and manipulate groups of data efficiently.**

---

## **1Ô∏è‚É£ Lists**
A lists is an ordered, changeable collection of items (duplicates allowed).
Think of a list like a **shopping list**: the order matters, and you can add/remove items as you go.

````python
# List example
fruits = ['apple', 'banana', 'orange', 'apple']
print(fruits)
print(type(fruits))
````

- `fruits` is a list.
- Lists are ordered, so the first item is 'apple', the second is 'banana', etc.
- Lists are mutable ‚Äî you can change them after creation:

````python
fruits.append('grape')  # add an item
print(fruits)
````

````python
fruits[1] = 'blueberry'  # change the SECOND item (indexing the list begins with zero; fruits[0] is the first item)
print(fruits)
````

## **2Ô∏è‚É£ Tuples**

A tuple is an ordered, unchangeable collection of items (duplicates allowed).
Tuples are like **lists you can‚Äôt change**. Think of coordinates on a map `(latitude, longitude)` ‚Äî once you set them, they don‚Äôt move.

The key for letting the program know you're assigning a variable multiple items as a tuple vs. a list is...use parentheses `()` instead of brackets`[]`-- brackets are for lists!

````python
# Tuple example
coordinates = (40.7128, -74.0060)
print(coordinates)
print(type(coordinates))
````

* Tuples are **ordered** (you can access by index)
* Tuples are **immutable** ‚Äî you cannot change their items:

```
coordinates[0] = 41  # ‚ùå This attempt to change the first item would would raise an error!
```
üí° Tip: use tuples when you want **data to stay fixed**.

Note: Tuples **and** lists can take mixed types (integers, strings, etc.)

````python
# ex: 
date = (2025, "09", 23)
print(date)
print(type(date))
````

## **3Ô∏è‚É£ Sets**
A set is an unordered, changeable collection of **unique** items ‚Äî duplicates are automatically removed. 
Order doesn‚Äôt matter.

````python
# Set example
numbers = {1, 2, 3, 2, 3, 4}
print(numbers)
print(type(numbers))
````

* Notice that repeated numbers disappeared? That‚Äôs because sets **only store unique values**.
* Items for lists must be enclosed in `()`; for tuples, in `[]`... and for sets, in `{}`.
* Sets are **mutable**, so you can add items:

````python
numbers.add(5)
print(numbers)
````

Remember: when adding an item to a **list**, instead of a **set**, we use the method `.append` rather than `.add`!  
(*A method is a function **that belongs** to a specific data type or object*)

## **4Ô∏è‚É£ Dictionaries**

Dictionaries store data in **key-value pairs**.  
You can think of a dictionary like a **contact book**: `"Alice" ‚Üí 555-1234`.

````python
# Dictionary example
student = {
    "name": "Maria",
    "age": 25,
    "grades": [88, 92, 79]
}
print(student)
print(type(student))
````

Then, you can access specific values using the **keys** (the first part of the item-pair)

````python
# example
print(student["grades"])
````

‚ö†Ô∏è Syntax gets a little trickier with dictionaries. Notice in the above example that:

- A dictionary is assigned using curly braces `{}`, just like sets ‚Äî but inside you store key‚Äìvalue pairs separated by colons.  
- Keys must follow the syntax rules of their data type (for example, string keys like `"name"` must be inside quotes).  
- When accessing values, you use indexing syntax, just like when accessing an item in a list. So use square brackets `[]` with the key inside.
