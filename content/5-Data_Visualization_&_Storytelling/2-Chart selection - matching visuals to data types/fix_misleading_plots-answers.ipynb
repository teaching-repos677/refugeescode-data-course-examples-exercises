```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Why are these plots misleading or bad?
# See if you can fix them
```

```python
flights = sns.load_dataset("flights")

sns.lineplot(data=flights, x="year", y="passengers")
plt.ylim(300, 600)  # misleading truncation
plt.title("Air Travel Growth Over Time")
plt.show()
```

```python
# Fixed 

flights = sns.load_dataset("flights")

sns.lineplot(data=flights, x="year", y="passengers")
plt.title("Air Travel Growth Over Time")
plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.barplot(data=tips, x="day", y="tip")
plt.ylim(2.5, 4)
plt.title("Average Tip by Day")
plt.show()
```

```python
# Notice differences appear bigger than they really are because the y axis starts at 2.5
```

```python
sns.barplot(data=tips, x="day", y="tip")
plt.title("Average Tip by Day")
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", bins=3)
plt.title("Distribution of Total Bills")
plt.show()
```

```python
# using too few or too many bins 
```

```python
sns.histplot(data=tips, x="total_bill", bins=20)
plt.title("Distribution of Total Bills")
plt.show()
```

```python
titanic = sns.load_dataset("titanic")

sns.countplot(data=titanic, x="class", hue="survived")
plt.title("Survival by Passenger Class")
plt.show()
```

```python

penguins = sns.load_dataset("penguins")

sns.lineplot(
    data=penguins,
    x="species",
    y="flipper_length_mm",
    estimator="mean"
)
plt.title("Average Flipper Length by Species")
plt.show()
```

```python
# Problem: Line Chart for Unordered Categories - better to use a bar plot
```

```python
sns.barplot(
    data=penguins,
    x="species",
    y="flipper_length_mm",
    estimator="mean"
)
plt.title("Average Flipper Length by Species")
plt.show()
```
