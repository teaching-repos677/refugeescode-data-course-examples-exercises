```python
# line graph - plot age and average height. Prepare the data:
import pandas as pd

data = [
    {"age": 1, "avg_height": 76},
    {"age": 2, "avg_height": 88},
    {"age": 3, "avg_height": 95},
    {"age": 4, "avg_height": 103},
    {"age": 5, "avg_height": 110},
    {"age": 6, "avg_height": 116},
    {"age": 7, "avg_height": 121},
    {"age": 8, "avg_height": 127},
    {"age": 9, "avg_height": 132},
    {"age": 10, "avg_height": 141},
    {"age": 11, "avg_height": 143},
    {"age": 12, "avg_height": 150},
    {"age": 13, "avg_height": 156},
    {"age": 14, "avg_height": 163},
]

df = pd.DataFrame(data)

df["max_height"] = df["avg_height"] + 5

df.head()
```

```python
# Make a basic line plot in matplotlib, using age on the x axis and avg_height on the y axis
```

```python
# Answer

plt.plot(df.age, df.avg_height);
```

```python
# Add title and axis labels
```

```python
# Answer
plt.plot(df.age, df.avg_height)
plt.title("Average Height by Age")
plt.xlabel("Age")
plt.ylabel("Average Height");
```

```python
# Add another line - max height, and a legend
```

```python
# Answer

plt.plot(df.age, df.avg_height, label="Average Height")
plt.plot(df.age, df.max_height, label="Max Height")

plt.title("Height by Age")
plt.xlabel("Age")
plt.ylabel("Height")
plt.legend()

plt.show()
```

```python
# Make a line chart in seaborn and plotly with the same data
```

```python
# Answer - Seaborn

sns.lineplot(data=df, x="age", y="avg_height")

plt.title("Average Height by Age")
plt.xlabel("Age")
plt.ylabel("Average Height")

plt.show()
```

```python
# Answer - Plotly
px.line(
    df,
    x="age",
    y="avg_height",
    title="Average Height by Age",
    labels={
        "age": "Age",
        "avg_height": "Average Height"
    }
)
```

```python
# From now on, we will use the tips dataset from seaborn 

tips = sns.load_dataset('tips')
tips.head()
```

# Bar plots

```python
# Make bar plots with day on the x axis and tip on the y axis in all three libraries. 
```

```python
# Answer - matplotlib 

plt.bar(tips.day, tips.tip);
```

```python
# Answer - seaborn

sns.barplot(data=tips, x='day', y='tip')
```

```python
# Answer - Plotly

px.bar(
    tips,
    x="day",
    y="tip",
    title="Average Tip by Day",
    labels={"tip": "Average Tip ($)", "day": "Day"}
)
```

# Histograms

```python
# Make histograms showing the distribution of total_bill in all three libraries
```

```python
# Answer - matplotlib

plt.hist(tips.total_bill)
plt.xlabel("total bill");
```

```python
# Answer - seaborn

sns.histplot(
    data=tips,
    x="total_bill",
    bins=20,
    kde=False
);
```

```python
# Answer - plotly

px.histogram(
    tips,
    x="total_bill",
    nbins=30,
    title="Distribution of Total Bill"
)
```
