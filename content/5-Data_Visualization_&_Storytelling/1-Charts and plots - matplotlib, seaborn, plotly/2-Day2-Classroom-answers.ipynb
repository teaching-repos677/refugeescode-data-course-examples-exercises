```python
# Activate your virtual environment - make sure you have visualization libraries installed

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Load the tips dataset

tips = sns.load_dataset('tips')
tips.head()
```

# Scatter plots

```python
# From now on, we will use the tips dataset from seaborn 

tips = sns.load_dataset('tips')
tips.head()
```

```python
# Make scatter plots using all three libraries. Put total_bill on the x axis and tip on the y axis
```

```python
# Answer - matplotlib

plt.scatter(data=tips, x='total_bill', y='tip')
```

```python
# Answer - seaborn

sns.scatterplot(data=tips, x='total_bill', y='tip', s=100, alpha=0.7)
```

```python
# Answer - plotly

px.scatter(
    tips,
    x="total_bill",
    y="tip",
#    color="sex",
    title="Tip vs Total Bill by Time of Day"
)
```

```python
# Answer - adding optional arguments 

fig = px.scatter(
    tips,
    x="total_bill",
    y="tip",
    color="sex",
    size="size",
    hover_data=["day", "time"],
    title="Tip vs Total Bill"
)
fig.show()
```

```python
# Add a hue of sex to the seaborn scatterplot. What changes?

sns.scatterplot(data=tips, x='total_bill', y='tip', s=100, alpha=0.7, hue="sex")
```

# Heatmaps

```python
#Correlation Heatmap

# Take only the numeric columns to find correlation
num_cols = list(tips.select_dtypes(include='number'))
```

```python
# Make correlation heatmaps in seaborn and plotly
```

```python
# Answer - heatmap in seaborn
sns.heatmap(data=tips[num_cols].corr(), 
            annot=True, cmap=sns.cubehelix_palette(as_cmap=True))
```

```python
# Answer - plotly

fig = px.imshow(
    tips[num_cols].corr(),
    text_auto=".2f",
    color_continuous_scale="RdBu",
    title="Correlation Heatmap"
)

fig.show()
```

# Boxplots and violin plots

```python
# How does the boxplot work?
```

```python
# Make boxplots in seaborn and plotly using day on the x axis and total_bill on the y axis. 
# Try experimenting with adding hue or color. 
```

```python
# Answer - boxplot in seaborn 

sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', palette='viridis')
```

```python
# Answer - boxplot in plotly

fig = px.box(
    tips,
    x="day",
    y="total_bill",
    title="Total Bill Distribution by Day",
    labels={
        "day": "Day",
        "total_bill": "Total Bill ($)"
    }
)

fig.show()
```

# Violin plots

```python
# Using the same data as with the boxplots, make violin plots in seaborn and plotly
```

```python
# Answer - seaborn 
sns.violinplot(data=tips, x='day', y='total_bill', palette='viridis')
```

```python
# Answer - plotly

fig = px.violin(
    tips,
    x="day",
    y="total_bill",
    box=True,       # show boxplot inside
    points="all",   # show individual points
    title="Total Bill Distribution by Day (Violin Plot)",
    labels={
        "day": "Day",
        "total_bill": "Total Bill ($)"
    }
)

fig.show()
```
